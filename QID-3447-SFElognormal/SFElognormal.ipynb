```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
normaxis = np.arange(-5, 15.1, 0.1)
lognormaxis = np.arange(0.01, 15.1, 0.1)
# Calculate the normal and log-normal distributions
n = norm.pdf(normaxis)
ln = norm.pdf(np.log(lognormaxis))
```

```python

# Plot
plt.figure()

# Plot the normal distribution
plt.plot(normaxis, n, color='blue', linestyle='-', linewidth=3, label='Normal')

# Add the log-normal distribution to the same plot
plt.plot(lognormaxis, ln, color='red', linestyle='--', linewidth=3, label='Log Normal')

# Set plot limits, title, and axis labels
plt.xlim(-5, 15)
plt.ylim(0, 0.5)
plt.title('Normal and Log Normal distribution')
plt.xlabel('X')
plt.ylabel('Y')

# Add a legend to identify the lines
plt.legend()

# Display the plot
plt.show()
```
