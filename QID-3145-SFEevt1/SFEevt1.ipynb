```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import genextreme

n = 100
sp = 5
xpos = sp * np.arange(1, n+1) / n
xneg = -sp + xpos
x = np.concatenate([xneg, xpos])
alpha = 1/2

fig, ax = plt.subplots(figsize=(10, 10))
fig.patch.set_alpha(0.0) 
ax.patch.set_alpha(0.0)   


# Gumbel (c=0)
gumb = np.column_stack([x, genextreme.pdf(x, c=0, loc=0, scale=0.5)])

# Frechet (c > 0)
frec = np.column_stack([x, genextreme.pdf(x, c=alpha, loc=0, scale=0.5)])

# Weibull (c < 0)
weib = np.column_stack([x, genextreme.pdf(x, c=-alpha, loc=0, scale=0.5)])

ax.plot(weib[:, 0], weib[:, 1], color='blue', linewidth=3, linestyle='--', label='Weibull')
ax.plot(frec[:, 0], frec[:, 1], color='red', linewidth=3, linestyle=':', label='Frechet')
ax.plot(gumb[:, 0], gumb[:, 1], color='black', linewidth=3, label='Gumbel')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Extreme Value Densities')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

#plt.savefig('extreme_value_densities.png', transparent=True, dpi=300, bbox_inches='tight')
plt.show()
```
