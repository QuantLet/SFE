```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 1000       # number of observations
a = 0.1        # mean reversion speed
sigma = 0.01   # volatility
b = 0.03       # long-run mean
delta = 0.1    # time step

# Simulate CIR process
x = np.zeros(n)
x[0] = b
for i in range(1, n):
    x[i] = x[i-1] + a * (b - x[i-1]) * delta + sigma * np.sqrt(delta * abs(x[i-1])) * np.random.normal()

# Plotting
index = np.arange(1, n + 1)
fig, ax = plt.subplots(figsize=(10, 5), facecolor='none')
ax.plot(index, x, color='blue')
ax.set_title('Simulated CIR Process')
ax.set_xlabel('i')
ax.set_ylabel('Values of the process')
ax.set_facecolor('none')  # Transparent background
plt.grid(False)
plt.show()
```
